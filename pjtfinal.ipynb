```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, GRU, Dropout, Input
from tensorflow.keras.callbacks import EarlyStopping

data = pd.read_csv('air_pollution_data.csv')


data['date'] = pd.to_datetime(data['date'], errors='coerce', dayfirst=True)
data = data.dropna(subset=['date'])


city_encoded = pd.get_dummies(data['city'], prefix='city', drop_first=True)


encoded_data = pd.concat([data, city_encoded], axis=1).drop(columns=['city'])


X = encoded_data.drop(columns=['aqi', 'date'])
y = encoded_data['aqi']


X = X.apply(pd.to_numeric, errors='coerce')


scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)


X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)


timesteps = 1
n_features = X_train.shape[1]
X_train_reshaped = X_train.reshape((X_train.shape[0], timesteps, n_features))
X_test_reshaped = X_test.reshape((X_test.shape[0], timesteps, n_features))



def build_lstm():
    model = Sequential([
        LSTM(units=50, return_sequences=True, input_shape=(timesteps, n_features)),
        Dropout(0.2),
        LSTM(units=50, return_sequences=False),
        Dropout(0.2),
        Dense(units=25),
        Dense(units=1)
    ])
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

def build_gru():
    model = Sequential([
        GRU(units=50, return_sequences=True, input_shape=(timesteps, n_features)),
        Dropout(0.2),
        GRU(units=50, return_sequences=False),
        Dropout(0.2),
        Dense(units=25),
        Dense(units=1)
    ])
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model


def build_transformer():
    model = Sequential([
        Dense(128, activation='relu', input_shape=(timesteps, n_features)),
        Dropout(0.2),
        Dense(64, activation='relu'),
        Dropout(0.2),
        Dense(1)
    ])
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model


def build_autoencoder():
    model = Sequential([
        Input(shape=(n_features,)),
        Dense(64, activation='relu'),
        Dropout(0.2),
        Dense(32, activation='relu'),
        Dense(64, activation='relu'),
        Dense(n_features, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model


def evaluate_model(model, X_train, y_train, X_test, y_test, model_name):
    early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
    model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=50, batch_size=32, callbacks=[early_stopping], verbose=1)
    y_pred = model.predict(X_test).flatten()
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    return {'Model': model_name, 'MSE': mse, 'MAE': mae, 'R2 Score': r2}


results = []


lstm_model = build_lstm()
results.append(evaluate_model(lstm_model, X_train_reshaped, y_train, X_test_reshaped, y_test, "LSTM"))


gru_model = build_gru()
results.append(evaluate_model(gru_model, X_train_reshaped, y_train, X_test_reshaped, y_test, "GRU"))

transformer_model = build_transformer()
results.append(evaluate_model(transformer_model, X_train_reshaped, y_train, X_test_reshaped, y_test, "Transformer"))


autoencoder = build_autoencoder()
X_train_flat = X_train_reshaped.reshape(X_train_reshaped.shape[0], -1)
X_test_flat = X_test_reshaped.reshape(X_test_reshaped.shape[0], -1)
autoencoder.fit(X_train_flat, X_train_flat, validation_data=(X_test_flat, X_test_flat), epochs=50, batch_size=32, verbose=1)
reconstructed = autoencoder.predict(X_test_flat)
reconstruction_error = np.mean(np.square(X_test_flat - reconstructed), axis=1)
mse_auto = np.mean(reconstruction_error)
results.append({'Model': 'Autoencoder', 'MSE': mse_auto, 'MAE': None, 'R2 Score': None})


results_df = pd.DataFrame(results)


print("\nModel Performance Comparison:")
print(results_df)


plt.figure(figsize=(10, 6))
metrics = ['MSE', 'MAE', 'R2 Score']
for metric in metrics:
    plt.plot(results_df['Model'], results_df[metric], marker='o', label=metric)
plt.title("Model Comparison Across Metrics")
plt.xlabel("Model")
plt.ylabel("Metric Value")
plt.legend()
plt.grid(True)
plt.show()


history = transformer_model.fit(X_train_reshaped, y_train, validation_data=(X_test_reshaped, y_test), epochs=50, batch_size=32, verbose=1)


for city in city_encoded.columns:
    city_name = city.split('_')[1]
    city_data = data[data['city'] == city_name]

    if city_data.empty:
        print(f"No data available for city: {city_name}")
        continue

    X_city = encoded_data[encoded_data[city] == 1].drop(columns=['aqi', 'date'])
    X_city_scaled = scaler.transform(X_city)
    X_city_reshaped = X_city_scaled.reshape((X_city_scaled.shape[0], timesteps, n_features))
    city_pred = transformer_model.predict(X_city_reshaped)


    city_data['prediction'] = city_pred.flatten()
    city_data['year'] = city_data['date'].dt.year
    historical_annual = city_data.groupby('year').agg({'aqi': 'mean', 'prediction': 'mean'}).reset_index()


    future_years = range(historical_annual['year'].max() + 1, historical_annual['year'].max() + 3)
    future_predictions = []
    for future_year in future_years:
        synthetic_input = np.mean(X_city_scaled, axis=0).reshape((1, timesteps, n_features))
        future_pred = transformer_model.predict(synthetic_input).flatten()[0]
        future_predictions.append({'year': future_year, 'prediction': future_pred})

    future_annual = pd.DataFrame(future_predictions)
    combined_annual = pd.concat([historical_annual[['year', 'aqi']].rename(columns={'aqi': 'value'}),
                                  future_annual.rename(columns={'prediction': 'value'})], ignore_index=True)


    plt.figure(figsize=(12, 6))


    plt.plot(
        historical_annual['year'],
        historical_annual['aqi'],
        label='Historical AQI',
        color='red',
        linestyle='-',
        linewidth=2,
        marker='o'
    )


    plt.plot(
        combined_annual['year'],
        combined_annual['value'],
        label='Predicted AQI',
        color='black',
        linestyle='--',
        linewidth=2,
        marker='o'
    )

    plt.title(f'AQI Prediction for {city_name} (Annual)', fontsize=16)
    plt.xlabel('Year', fontsize=12)
    plt.ylabel('AQI', fontsize=12)
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()
```

```python
def generate_aqi_description(historical_annual, future_predictions, city_name):
    # Extract the most recent year and AQI values for description generation
    last_year = historical_annual['year'].max()
    historical_aqi = historical_annual[historical_annual['year'] == last_year]['aqi'].values[0]

    # Get the forecasted AQI for the next two years
    future_aqi = future_predictions[0]['prediction']  # for next year
    future_aqi_2 = future_predictions[1]['prediction']  # for 2 years ahead

    # Create an input prompt for the model
    prompt = (f"In {last_year}, {city_name} experienced an average AQI of {historical_aqi}, indicating "
              f"pollution levels. Forecasts suggest an improvement in the next two years, with predicted AQI "
              f"of {future_aqi} for {last_year + 1} and {future_aqi_2} for {last_year + 2}. "
              f"Describe the AQI trends for {city_name}.")

    # Generate description using the Hugging Face model
    description = generator(prompt, max_length=150, num_return_sequences=1)[0]['generated_text']
    return description
```
